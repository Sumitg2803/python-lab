```python
#write a program to generate and handle division by zero error by using try

def aby(a, b):
    try:
        c=(a+b)/(a-b)

    except ZeroDivisionError:
        print("Division by zero occurred!")
    else:
        print("The result is:", c)
aby (1, 11)
aby (4, 2)
```
